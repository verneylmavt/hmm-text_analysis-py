```python
x_all = pd.DataFrame(token_all, columns=['Token'])

# Need to check more, this takes >1m to do
new_x_all = pd.DataFrame({'Token': np.repeat(x_all['Token'], 7)}).reset_index()
new_x_all.columns = ['Index', 'Token']
new_x_all["Label"] = np.tile(label_all, int(len(new_x_all)/7))
count_x_all = []
location_x_all = []

for i in token_all:
    for j in label_all:
        cond_1 = arr_token == i
        cond_2 = arr_label == j
        loc = (np.where(cond_1 & cond_2)[0]).tolist()
        # df_ij = df.iloc[loc, :]
        count_x_all += [len(loc)]
        location_x_all += [loc]
new_x_all["Count"] = count_x_all
new_x_all["Location"] = location_x_all

print(len(new_x_all))
print(type(new_x_all))
display(new_x_all)
```

$1. \:\:\: e(x \mid y)=\frac{\operatorname{Count}(y \rightarrow x)}{\operatorname{Count}(y)}$

```python
def emm_par(x, y):
    # x is an instance of Token
    # y is an instance of Label
    cond_1 = new_x_all["Token"] == x
    cond_2 = new_x_all["Label"] == y
    num = ((new_x_all[cond_1 & cond_2])["Count"]).values[0]
    cond_3 = y_all["Label"] == y
    denom = ((y_all[cond_3])["Count"]).values[0]
    return num/denom
```

```python
def calculate_emission_params(token_label_list):
    emission_params = {}
    for x, y in token_label_list:
        emission_params[(x, y)] = emm_par(x, y)
    return emission_params
```

```python
emission_params_dict=calculate_emission_params(token_label_list)
print(emission_params_dict)
```

```python
a = emm_par("Estuvimos", "O")
print(a)

b = emm_par("estuvimos", "O")
print(b)
```

$2. \:\:\:e(x \mid y)=\left\{\begin{array}{cc}
\frac{\operatorname{count}(y \rightarrow x)}{\operatorname{Count}(y)+k} & \text { If the word token } x \text { appears in the training set } \\
\frac{k}{\operatorname{Count}(y)+k} & \text { If word token } x \text { is the special token \#UNK\# }
\end{array}\right.$

```python
def emm_par_test(x, y):
    # x is an instance of Token
    # y is an instance of Label
    k = 1
    if x in new_x_all.loc[:, "Token"].unique():
        cond_1 = new_x_all["Token"] == x
        cond_2 = new_x_all["Label"] == y
        num = ((new_x_all[cond_1 & cond_2])["Count"]).values[0]
    else:
        num = k
    cond_3 = y_all["Label"] == y
    denom = ((y_all[cond_3])["Count"]).values[0] + k
    return num/denom
```

```python
a = emm_par_test(",", "I-negative")
print(a)

b = emm_par_test("Kontollllllllllll", "I-neutral")
print(b)
```

$3. \:\:\:y^*=\underset{y}{\arg \max } \:e(x \mid y)$

```python
df_y = pd.DataFrame()

emm_par_score_all = []
y_star_all = []
token_all = np.append(token_all, "#UNK#")
print(token_all)
```

```python
# Need to check more, this takes >2m to do
for i in token_all:
    emm_par_score = 0
    y_star = ""
    for j in label_all:
        emm_par_score_temp = emm_par_test(i,j)
        if emm_par_score_temp > emm_par_score:
            emm_par_score = emm_par_score_temp
            y_star = j
        else:
            pass
    y_star_all += [y_star]
    emm_par_score_all += [emm_par_score]
```

```python
df_y["Token"] = token_all
df_y["y*"] = y_star_all
df_y["Max e(x|y)"] = emm_par_score_all

display(df_y)

for i in label_all:
    print(emm_par_test("sin", i))
```

```python
def estimate_transition_parameters(labels):
    """
    Estimates the transition parameters from a sequence of labels using Maximum Likelihood Estimation (MLE).

    Args:
        labels (list): A list of labels representing a sequence of tags.

    Returns:
        dict: A dictionary containing transition probabilities as {('previous_tag', 'current_tag'): probability}.
    """

    # Count the occurrences of each (previous_tag, current_tag) pair
    transition_counts = {}
    

    

    for i in range(len(labels) - 1):
        prev_tag = labels[i]
        current_tag = labels[i + 1]
        # Increment the count of the current_tag given the previous_tag
        transition_counts[(prev_tag, current_tag)] = transition_counts.get((prev_tag, current_tag), 0) + 1

    
    # transition_counts.pop(next(iter(transition_counts)))
    # transition_counts.popitem()
    # transition_counts.pop(next(iter(transition_counts)))
    # transition_counts.popitem()

    # Calculate transition probabilities based on MLE
    transition_probabilities = {}
    for (prev_tag, current_tag), count in transition_counts.items():
        # Calculate the probability q(current_tag | prev_tag) using MLE formula
        prev_tag_count = labels_list.count(prev_tag)
        transition_probabilities[( prev_tag,current_tag)] = count / prev_tag_count

    return transition_probabilities

```

```python

transition_params_dict=estimate_transition_parameters(labels_list)
print(transition_params_dict)
```

```python
def viterbi(obs, states,  transmission_dict,emission_dict):
    viterbi = [{}]
    path = {}

    # Initialize base cases
    for state in states:        
        if state=="START":
            viterbi[0][state] = 1
        else:
            viterbi[0][state] = 0
            

        
        path[state] = [state]

    # Viterbi algorithm
    for t in range(1, len(obs)):
        viterbi.append({})
        new_path = {}

        for state  in states:
            max_prob = max(viterbi[t-1][prev_state] * transmission_dict.get((prev_state,state),0) * emission_dict.get((obs[t],state),0)   for prev_state in states)
            viterbi[t][state] = max_prob
            new_path[state] = path[max(path, key=lambda x: viterbi[t-1][x] * transmission_dict.get((x,state),0))]

        path = new_path

    # Find the last state with maximum probability
    last_state = max(viterbi[-1], key=lambda x: viterbi[-1][x])

    # Return the most likely sequence of states
    return path[last_state]
# Usage example
sentence = ['Estuvimos', 'hace', 'poco']
transition_params_dict = {('O', 'I-PER'): 0.2, ('I-PER', 'I-PER'): 0.5, ('I-PER', 'O'): 0.5, ('O', 'O'): 0.4, ('O', 'I-LOC'): 0.2, ('I-LOC', 'I-LOC'): 0.5, ('I-LOC', 'O'): 0.5}
emission_params_dict = {('Estuvimos', 'O'): 0.0002066471499913897, ('hace', 'O'): 0.0008954709832960221, ('poco', 'O'): 0.0018942655415877389}

path = viterbi(sentence,labels_list, transition_params_dict, emission_params_dict)
print(path)
```
